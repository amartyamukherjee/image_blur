```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('jerry.jpg')
```

```python
b,g,r = cv2.split(img)
rgb_img = cv2.merge([r,g,b])
plt.xticks([])
plt.yticks([])
plt.imshow(rgb_img)
```

```python
plt.show()
```

```python
r_blur = r
g_blur = g
b_blur = b
```

```python
k = 1        #Constant for heat equation
dt = 0.1       #finite difference approximation in time
n = 10         #number of iterations
```

```python
for t in range(n):
    for i in range(1, len(r)-1):
        for j in range(1, len(r[0])-1):
            r_yy = (r[i-1][j]+r[i+1][j]-2*r[i][j])/4      #find second derivatives for each color using
            r_xx = (r[i][j-1]+r[i][j+1]-2*r[i][j])/4      #finite difference approximation
            r_blur[i][j] = r[i][j]+k*dt*(r_xx+r_yy)
            g_yy = (g[i-1][j]+g[i+1][j]-2*g[i][j])/4
            g_xx = (g[i][j-1]+g[i][j+1]-2*g[i][j])/4
            g_blur[i][j] = g[i][j]+k*dt*(g_xx+g_yy)
            b_yy = (b[i-1][j]+b[i+1][j]-2*b[i][j])/4
            b_xx = (b[i][j-1]+b[i][j+1]-2*b[i][j])/4
            b_blur[i][j] = b[i][j]+k*dt*(b_xx+b_yy)
    r = r_blur
    g = g_blur
    b = b_blur
```

```python
rgb_img = cv2.merge([r_blur,g_blur,b_blur])
plt.imshow(rgb_img)
plt.imshow(rgb_img)
plt.xticks([])
plt.yticks([])
```

```python
plt.show()
```

```python
cv2.imwrite('jerry_blurred.jpg',cv2.merge([b_blur,g_blur,r_blur]))
```
